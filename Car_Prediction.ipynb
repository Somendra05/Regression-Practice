### Encoding categorical data

```python
df.replace({'Fuel_Type': {'Petrol' : 0, 'Diesel' : 1, 'CNG' : 2 }}, inplace = True)

df.replace({'Seller_Type': {'Dealer' : 0, 'Individual' : 1}}, inplace = True)

df.replace({'Transmission' : {'Manual' : 0, 'Automatic' : 1}}, inplace = True)
```

```python
df.head()
```

### Spiltting the data and target variable

Selling_price is the target feature so we will remove it.

We will remove car name as well as its not needed

```python
X = df.drop(['Car_Name', 'Selling_Price'], axis=1)
Y = df['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

### Spliting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, random_state = 2)
```

### Model Training

1. Linear Regression

```python
linear_regression = LinearRegression()
```

```python
linear_regression.fit(X_train, Y_train)
```

Model Evaluation

```python
training_data_prediction = linear_regression.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
error_score
```

Visualizing the actual and predicted price

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```

```python
test_data_prediction = linear_regression.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction)
error_score
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```

2. Lasso Regression

```python
lass_reg_model = Lasso()
```

```python
lass_reg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
training_data_prediction = lass_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = lass_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```
